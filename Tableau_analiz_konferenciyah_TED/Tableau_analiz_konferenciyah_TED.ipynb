# Проект  Tableau Public  основе данных о конференциях TED

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
**Описание проекта:**

TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы. 
В разное время на TED-конференциях выступали математик Бенуа Мандельброт, теоретик искусственного интеллекта Марвин Минский, спортсменка Дана Ньяд и основатель Google Ларри Пейдж. В истории TED также были неоднозначные и даже скандальные выступления. Например, в 2010 году на конференции выступил Рэнди Пауэлл с рассказом о псевдонаучной «вихревой математике», а в 2014 году в конференции TEDMED участвовала Элизабет Холмс — основательница печально известного стартапа Theranos.
В этом проекте вы исследуете историю TED-конференций с помощью Tableau.
</div>

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
**Описание данных:**
    
Данные выступленииях
    
* talk_id — идентификатор выступления;
* url — ссылка на запись выступления;
* title — название выступления;
* description — краткое описание;
* film_date — дата записи выступления;
* duration — длительность в секундах;
* views — количество просмотров;
* main_tag — основная категория, к которой относится выступление;
* speaker_id — уникальный идентификатор автора выступления;
* laughter_count — количество раз, когда аудитория смеялась в ходе выступления;
* applause_count — количество раз, когда аудитория аплодировала в ходе выступления;
* language — язык, на котором велось выступление;
* event_id — уникальный идентификатор конференции.
    
Справочник конференций
    
* conf_id — уникальный идентификатор конференции;
* event — название конференции;
* country — страна проведения конференции.
    
Справочник авторов выступления
    
* author_id — уникальный идентификатор автора выступления;
* speaker_name — имя автора;
* speaker_occupation — профессиональная область автора;
* speaker_description — описание профессиональной деятельности автора.

## Дашборд «История выступлений»

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
Самыми активными спикерами на TED являются авторы из США -55%, на втором месте авторы из Великобритании - почти 16%, на третьем месте авторы из Канады -почти 11%.  
    Наиболее популярные категории выступлений наука, технологии, культура, общество.  
    В 2020-2021 году резко упало количество выступлений и на первых позициях оказались категории общество и его изменение, глобальные вопросы и технологии. 
</div>

## Дашборд «Тематики выступлений»

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
Самые популярные категории выступлений: наука -19,12%, технологии – 15,23%, культура, культура – 11,85%.   
    В Топ-3 стран по количеству авторов из данных стран США, Великобритания, Канада. Лидирующие позиции по категориям выступлений совпадают с общим трендом (Наука, технологии, культура), а так же между этими Топ странами. Мы можем отметить, что категория наука для Канады более популярна, чем для Великобритании и США. А также, что категория культура занимает в Великобритании.  
    Больше всего аплодисментов собирают выступления в категории: искусство, общество и образование. Реже из топ10 категорий получают выступления тематики: наука и глобальные изменения.  
    Самое большое количество аплодисментов получило выступления. The political progress women have made — and what's next / Cecile Richards.  
    Аудитория сочла самыми забавными выступления из категорий: искусство, образование и дизайн. Меньше всего смеха было в категориях: технологии и глобальные изменения.  
    Самое большое количество смешных моментов по мнению аудитории у видео Did you hear the one about the Iranian-American? / Maz Jobrani.  
    Мы видим, что больше нет прямой взаимосвязи между длительностью выступления и количеством просмотров. Большинство выступлений длиться от 3 до 22 минут и набирают до 5 миллионов просмотров. Самое длинное из представленных видео длиться 62,5 минуты - 3 secrets to Netflix's success / Reed Hastings and Chris Anderson. Наибольше количество просмотров у видео Your body language may shape who you are | Amy Cuddy – 21 миллион просмотров.

## Дашборд «Авторы выступлений»

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
Преобладающими областями деятельности авторов являются: писатель, журналист, артист.  
    Чаще всего на одного автора приходится одно выступление. Однако есть авторы, которые выступали не один раз. Hans Rosling выступал восемь раз, Marco Tempest - семь.  
    Среди дизайнеров больше всех выступал Tom Wujec, который провел 4 выступления. Среди всех его выступлений, речь на тему Learn to use the 13th-century astrolabe посмотрели меньше всего (940 542 просмотров). 
</div>

## Дашборд «Популярность TED выраженная в просмотрах»

<div  style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
    
Самыми успешными для TED стали 2016 и 2017 годы. В эти года конференции собирали больше всего просмотров. В 2020 резкое снижение популярности, из -за снижения количества выступлений.  
    Самые популярные за все время конференции: TED-2015, TED-2016, TED-2017. Это общие конференции без специфики по категориям.  В Топ странах зрители больше всего любят категории: наука, культура, технологии общество, что соответствует общему тренду.   
    Удивительным стало то, что просмотры у неврологов высокие, хотя они не вошли в топ профессиональных областей авторов. На втором месте писатели, которые как раз входят в топ.
